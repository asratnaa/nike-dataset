```python
import pandas as pd
```

```python
df = pd.read_csv('D:\jupyter\pandas\data_nike.csv')
```

```python
df.info()
```

```python
df
```

```python
df["Invoice Date"] = pd.to_datetime(df["Invoice Date"],errors="coerce")
```

```python
df["Sales Method"].unique()
```

```python
df_product = df.groupby(['Retailer'])
df_total = df_product.agg({'Units Sold':'sum','Total Sales':'sum'})
df_total.sort_values(by='Units Sold',ascending=False)
```

```python
df_sales = df.groupby(["Sales Method"])
df_sales_method = df_sales.agg({'Units Sold':'sum','Total Sales':'sum'})
df_sales_method.sort_values(by= "Units Sold", ascending=False)
```

```python
filter_1 = df["Sales Method"] == "Online"
df_filtered = df[filter_1]
df_product_1 = df_filtered.groupby(['Product'])
product_online = df_product_1.agg({'Units Sold':'sum','Total Sales':'sum'})
product_online.sort_values(by= "Units Sold", ascending=False)
```

```python
filter_sales = df["Sales Method"] == "Outlet"
sales = df[filter_sales]
sales_group = sales.groupby(["Product"])
sales_outlet = sales_group.agg({"Units Sold":"sum",'Total Sales':'sum'})
sales_outlet.sort_values(by="Units Sold",ascending=False)
```

```python
sales = df.groupby(["Product","Sales Method"])
sales_total = sales.agg({"Units Sold":"sum",'Total Sales':'sum'})
sales_total
```

```python
df["Year"] = df["Invoice Date"].dt.year
```

```python
sales_2020 = df[df['Year'] == 2020]
filter_2 = sales_2020.groupby(["Product","Sales Method"])
sal_2020 = filter_2.agg({"Units Sold":"sum",'Total Sales':'sum'})
sal_2020.sort_values(by= "Units Sold", ascending=False)
```

```python
sales_2021 = df[df['Year'] == 2021]
filter_2 = sales_2021.groupby(["Product","Sales Method"])
sal_2021 = filter_2.agg({"Units Sold":"sum",'Total Sales':'sum'})
sal_2021.sort_values(by= "Units Sold", ascending=False)
```
